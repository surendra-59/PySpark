### Topic 4: Displaying Data

```python
df = spark.read.csv("/Volumes/workspace/default/practice-data/employees/", header=True)
```

```python
# 1. show()
df.show()
```

```python
# 2. Show specific number of rows
df.show(5)
```

```python
# 3. Show without truncation
df.show(truncate=False)
```

```python
# 4. Show specific rows without truncation
df.show(10, truncate=False)
```

```python
# 5. Vertical display : Useful for debugging
df.show(10, vertical=True)
```

```python
# 6. Show all rows: Not recommended for huge datasets.
df.show(df.count())
```

```python
# 7. display() (Microsoft Fabric / Databricks Notebook)
# Note: display() is available in Microsoft Fabric notebooks and Databricks notebooks, but it is not a standard PySpark API. If you write pure PySpark code outside notebook environments, use show() instead.

display(df)
```

```python
# 8. Convert to Pandas (Small Data Only)

# Avoid this for large datasets because all data is collected into the driver's memory.

df.toPandas()
```
