### Topic 3: Creating Dataframe

What is a DataFrame?

A DataFrame is a table with rows and columns.
Just like an Excel sheet    
Almost every Spark operation starts with a DataFrame.

#### Ways to create dataframe:
There are several ways.

From Python List ⭐⭐⭐⭐⭐         
From List of Dictionaries ⭐⭐⭐⭐  
From Rows (Row object) ⭐⭐⭐   
From RDD (Rare)             
From Pandas DataFrame ⭐⭐          
Reading CSV/Parquet/Delta (already covered separately)

```python
# Method 1 — Create DataFrame from List of Tuples
data = [
    (1,"John", 50000),
    (2,"Alice", 65000),
    (3, "Bob", 70000)
]

# now define column names
columns = ["EmployeeID", "Name", "Salary"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Syntax: spark.createDataFrame(data, schema)
# data: Actual row,
# schema: can be list of column names or StructType(full schema with data types)
# Sparks now understand rows, columns, schema

# Display
df.show()
```

```python
# Method 2 — Without Column Names
data = [
    (1,"John"),
    (2,"Alice")
]

df = spark.createDataFrame(data)
df.show()

# Here spark create default name.
# _1, _2, _3, _4, _5, _6, _7, _8, _9, _10, _11, _12, _13,
```

```python
# Method 3 — List of Dictionaries

data = [
    {"id":1, "name":"John", "salary":50000},
    {"id":2, "name":"Alice", "salary":65000},
    {"id":3, "name":"Bob", "salary":65000}
]

df = spark.createDataFrame(data)
df.show()
```

```python
# Method 4 — Using Row Object

from pyspark.sql import Row

employees = [
    Row(id=1, name="John", salary=50000),
    Row(id=2, name="Alice", salary=65000),
    Row(id=3, name="Bob", salary=70000)
]

df = spark.createDataFrame(employees)
df.show()

# Why use Row?

# It creates named fields instead of positional values and is useful when building records programmatically.

# In short, tuples are just dumb containers of values. Row objects are smart, self-describing records that make your code readable, maintainable, and less prone to "index out of range" bugs.
```

```python
# Method 5 — From Pandas

import pandas as pd

pdf = pd.DataFrame({
    "id":[1,2],
    "name":["John", "Alice"]
})

# conver
df = spark.createDataFrame(pdf)
df.show()
df.printSchema()
```

## PySpark DataFrame Creation – Interview Questions

1. **What is a DataFrame?**
   → A distributed table with rows and columns.

2. **Which function creates a DataFrame?**
   → `spark.createDataFrame()`.

3. **What are the parameters of `createDataFrame()`?**
   → `data` and `schema` (or column names).

4. **Can Spark create a DataFrame from a Python list?**
   → Yes.

5. **Can Spark create a DataFrame from a list of tuples?**
   → Yes.

6. **Can Spark create a DataFrame from a list of dictionaries?**
   → Yes.

7. **Can Spark create a DataFrame from a Pandas DataFrame?**
   → Yes.

8. **Can Spark create a DataFrame from Row objects?**
   → Yes.

9. **What happens if column names are not provided?**
   → Spark assigns `_1`, `_2`, `_3`, etc.

10. **Which method displays the DataFrame?**
    → `df.show()`.

11. **Which method displays the schema?**
    → `df.printSchema()`.

12. **What is schema?**
    → The structure of a DataFrame (column names and data types).

13. **Does Spark infer data types automatically?**
    → Yes.

14. **What data type is inferred for `50000`?**
    → `LongType`.

15. **What data type is inferred for `50000.5`?**
    → `DoubleType`.

16. **What data type is inferred for `"John"`?**
    → `StringType`.

17. **Can a DataFrame have duplicate rows?**
    → Yes.

18. **Can a DataFrame have duplicate column names?**
    → No (not recommended).

19. **What error occurs if row lengths differ?**
    → Schema mismatch error.

20. **What error occurs if column names are fewer than values?**
    → Schema mismatch error.

21. **Is a DataFrame mutable?**
    → No, it is immutable.

22. **Can you modify an existing DataFrame directly?**
    → No, transformations create a new DataFrame.

23. **Which object is required to create a DataFrame?**
    → `SparkSession`.

24. **What is the default nullable property of inferred columns?**
    → `true`.

25. **Which is the most common way to create a DataFrame in interviews?**
    → `spark.createDataFrame(data, columns)`.

26. **What is a Row object?**
    → A named record used to create DataFrames.

27. **Why use Row objects?**
    → To define named fields instead of positional values.

28. **Can Spark infer column names from dictionaries?**
    → Yes.

29. **Which data structure is most commonly used to create a DataFrame?**
    → List of tuples.

30. **Is `createDataFrame()` an action or transformation?**
    → Neither; it creates a DataFrame.
