#### Part 1 — String Functions

```python
from pyspark.sql import functions as F

data = [
    (1, "Surendra Raj", "surendra@gmail.com", "kathmandu"),
    (2, "Alice Smith", "ALICE@gmail.COM", "Pokhara"),
    (3, "bob johnson", "bob@yahoo.com", "Lalitpur"),
    (4, None, None, "Bhaktapur")
]

columns = ["id", "name", "email", "city"]

df = spark.createDataFrame(data, columns)

df.show(truncate=False)
```

```python
# 1. upper()

df.select(
    "name",
    F.upper("name").alias("upper_name")
).show()
```

```python
# 2. lower()

df.select(
    "email",
    F.lower("email").alias("lower_email")
)
```

```python
# 3. initcap()
# Capitalizes the first letter of each word.

df.select(
    "name",
    F.initcap("name").alias("proper_name")
).show()
```

```python
# 4. trim()
    # Removes spaces from both sides.

df.select(
    "name",
    F.trim("name").alias("trim_name")
).show()
```

```python
# 5. ltrim()
   # Remove left spaces.

df.select(
    F.ltrim("name").alias("left_trim")
).show()
```

```python
# 6. rtrim()
    # Remove right spaces.

df.select(
    F.rtrim("name").alias("right_trim")
).show()
```

```python
# 7. length()

df.select(
    "name",
    F.length("name").alias("length")
).show()
```

```python
# 8. concat()
    # Join columns.

df.select(
    F.concat("name", F.lit(" - "), "city").alias("details")
).show(truncate=False)

# o/p
# +------------------------+
# |details                 |
# +------------------------+
# |Surendra Raj - kathmandu|
# |Alice Smith - Pokhara   |
# |bob johnson - Lalitpur  |
# |NULL                    |
# +------------------------+
```

```python
# 9. concat_ws()
    # Join using separator.

df.select(
    F.concat_ws(", ", "name", "city").alias("info")
).show(truncate=False)

# o/p: Surendra Raj, Kathmandu
```

```python
# 10. substring()

df.select(
    "name",
    F.substring("name", 1, 5).alias("first5")
).show()

# o/p: 
# Suren
# Alice
# bob j
```

```python
# 11. regexp_replace()
    # Replace characters. : very common in DP-700 exam

df.select(
    F.regexp_replace("email", "@", " at ").alias("email")
).show(truncate=False)

# +---------------------+
# |email                |
# +---------------------+
# |surendra at gmail.com|
# |ALICE at gmail.COM   |
# |bob at yahoo.com     |
# |NULL                 |
# +---------------------+

df.select(
    F.translate("email", "@", "#").alias("new_email")
).show(truncate=False)

# +------------------+
# |new_email         |
# +------------------+
# |surendra#gmail.com|
# |ALICE#gmail.COM   |
# |bob#yahoo.com     |
# |NULL              |
# +------------------+


# There is a function named replace() in newer Spark versions, but its signature is:
#     replace(column, search, replace)
```

```python
# 12. split()

df1 = spark.createDataFrame(
    [("apple,banana,manago",)],
    ["fruits"]
)

df1.select(
    F.split("fruits", ",").alias("fruits_list")
).show(truncate=False)


# +-----------------------+
# |fruits_list            |
# +-----------------------+
# |[apple, banana, manago]|
# +-----------------------+
```

```python
# 13. contains()

df.filter(
    F.col("email").contains("gmail")
).show()

# +---+------------+------------------+---------+
# | id|        name|             email|     city|
# +---+------------+------------------+---------+
# |  1|Surendra Raj|surendra@gmail.com|kathmandu|
# |  2| Alice Smith|   ALICE@gmail.COM|  Pokhara|
# +---+------------+------------------+---------+
```

```python
# 14. startwith()

df.filter(
    F.col("city").startswith("K")
).show()
```

```python
# 15. endswith()

df.filter(
    F.col("email").endswith(".com")
).show()
```
