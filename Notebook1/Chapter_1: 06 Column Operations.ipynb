##### 1. Selecting Columns (select())
This is one of the most frequently used DF operations.

```python
# Sample Data
from pyspark.sql import Row

data = [
    Row(id=1, name="Alice", age=25, salary=50000),
    Row(id=2, name="Bob", age=30, salary=60000),
    Row(id=3, name="Charlie", age=35, salary=90000)
]

df = spark.createDataFrame(data)

df.show()
```

```python
# Select One Column
df.select("name").show()
```

```python
# Select Multiple Columns
df.select("name", "salary").show()
```

```python
#Using col()
    # Recommended in production code. It Avoids ambiguity between column names and Python variables

from pyspark.sql.functions import col

df.select(col("name"), col("salary")).show()
```

```python
# Select All Columns
df.select("*").show()
```

```python
# Selecting with Expressions
df.select(
    col("name"),
    col("salary") * 2
).show()
```

```python
# Alias
df.select(
    col('name'),
    (col('salary')*2).alias("double_salary")
).show()
```

##### 2. Renaming Columns (withColumnRenamed() )
Used very often after reading CSV files.


```python
# current schema
df.printSchema()
```

```python
# Rename one column
df.withColumnRenamed("salary", "monthy_salary").show()
```

```python
# Rename multiple columns
df = (
    df
    .withColumnRenamed("name", "employee_name")
    .withColumnRenamed("salary", "monthly_salary")
)

df.show()
```

#####3. Adding New Columns (withColumn())


```python
# Add Constant Value
from pyspark.sql.functions import lit

df.withColumn("country", lit("Nepal")).show()
```

```python
# Add Calculated Column
from pyspark.sql.functions import col

df.withColumn(
    "bonus",
    col("monthly_salary") * 0.10
).show()
```

```python
# Add Two Columns Together
df.withColumn(
    "total_salary",
    col("monthly_salary") + col("monthly_salary") * 0.10
).show()
```

```python
df.show()
```

```python
# Replace Existing Column
# withColumn() also replaces an existing column if the name matches.

df.withColumn(
    "monthly_salary",
    col("monthly_salary") + 1000
).show()
```

```python
##### 4. Dropping Columns (drop())
```

```python
# Drop one column
df.drop("age").show()
```

```python
# df.drop("age", "monthly_salary").show()

# Remember if you wanna delete permenently use
df = df.drop("age", "monthly_salary")
```

### PySpark Interview Questions - Selecting, Renaming, Adding & Dropping Columns

##### Q1. What is the purpose of `select()` in PySpark?
**Answer:** `select()` is used to retrieve one or more columns from a DataFrame.

Example:
python
df.select("name", "salary")


---

##### Q2. What is the difference between `select()` and `withColumn()`?
**Answer:**
- `select()` selects existing columns or expressions.
- `withColumn()` adds a new column or replaces an existing column.

---

##### Q3. How do you select multiple columns?
**Answer:**

python
df.select("name", "age", "salary")


---

##### Q4. What is the advantage of using `col()` instead of column names as strings?
**Answer:** `col()` is more explicit, improves readability, and is preferred when building expressions.

Example:

python
from pyspark.sql.functions import col

df.select(col("salary"))


---

##### Q5. What does `alias()` do?
**Answer:** `alias()` gives a temporary name to a selected column or expression.

Example:

python
df.select((col("salary") * 2).alias("double_salary"))


---

##### Q6. How do you rename a column?
**Answer:**

python
df.withColumnRenamed("salary", "monthly_salary")


---

##### Q7. Can you rename multiple columns?
**Answer:** Yes, by chaining `withColumnRenamed()`.

Example:

python
df.withColumnRenamed("name", "employee_name") \
  .withColumnRenamed("salary", "monthly_salary")


---

##### Q8. What is `withColumn()` used for?
**Answer:** It is used to add a new column or replace an existing column.

Example:

python
df.withColumn("bonus", col("salary") * 0.10)


---

##### Q9. Does `withColumn()` modify the original DataFrame?
**Answer:** No. DataFrames are immutable. It returns a new DataFrame.

---

##### Q10. What happens if the column already exists in `withColumn()`?
**Answer:** The existing column is replaced with the new values.

Example:

python
df.withColumn("salary", col("salary") + 1000)


---

##### Q11. How do you create a column with a constant value?
**Answer:** Use `lit()`.

Example:

python
from pyspark.sql.functions import lit

df.withColumn("country", lit("Nepal"))


---

##### Q12. What is `lit()`?
**Answer:** `lit()` creates a constant (literal) value that can be used as a DataFrame column.

---

##### Q13. How do you create a calculated column?
**Answer:**

python
df.withColumn("bonus", col("salary") * 0.10)


---

##### Q14. How do you combine multiple columns into a new column?
**Answer:**

python
df.withColumn(
    "total_salary",
    col("salary") + col("bonus")
)


---

##### Q15. How do you remove a column from a DataFrame?
**Answer:**

python
df.drop("age")


---

##### Q16. How do you drop multiple columns?
**Answer:**

python
df.drop("age", "salary")


---

##### Q17. Does `drop()` modify the original DataFrame?
**Answer:** No. It returns a new DataFrame.

---

##### Q18. How can you check the column names of a DataFrame?
**Answer:**

python
df.columns


---

##### Q19. How can you view the DataFrame schema?
**Answer:**

python
df.printSchema()


---

##### Q20. What are the most commonly used column transformation functions in PySpark?
**Answer:**
- `select()`
- `col()`
- `alias()`
- `withColumn()`
- `withColumnRenamed()`
- `drop()`
- `lit()`
