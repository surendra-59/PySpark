```python
from datetime import datetime, timedelta
import random

first_names = [
    "Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Henry",
    "Ivy", "Jack", "Kevin", "Laura", "Mike", "Nancy", "Oscar", "Paul",
    "Queen", "Ryan", "Sophia", "Tom", "Halland","Brook","Mike","Suresh","Ravin"
]

last_names = [
    "Smith", "Johnson", "Brown", "Williams", "Jones", "Miller",
    "Davis", "Wilson", "Moore", "Taylor","Cliff","Floki","Terry","Kalf"
]

departments = [
    "HR", "IT", "Finance", "Marketing",
    "Sales", "Operations", "Legal", "R&D"
]

cities = [
    "New York", "Chicago", "Boston", "Seattle",
    "Austin", "Dallas", "Miami",
    "Denver", "Los Angeles", "San Francisco"
]

start_date = datetime(2020, 1, 1)

employees_data = []

for emp_id in range(101, 201):

    hire_date = start_date + timedelta(days=random.randint(0, 2000))

    employees_data.append(
        (
            emp_id,
            random.choice(first_names),
            random.choice(last_names),
            random.choice(departments),
            random.randint(22, 60),
            random.randint(35000, 150000),
            random.choice(cities),
            hire_date.strftime("%Y-%m-%d")
        )
    )
```

```python
product_names = [
    "Laptop","Mouse","Keyboard","Monitor","Printer",
    "Desk","Chair","Lamp","Notebook","Pen",
    "Backpack","Bottle","Tablet","Phone","Headphones",
    "Speaker","Camera","Watch","Router","SSD",
    "USB Drive","Projector","Scanner","Webcam","Microphone",
    "Whiteboard","Marker","Stapler","Calculator","Folder",
    "Sofa","Cabinet","Bookshelf","Fan","Heater",
    "Football","Basketball","Cricket Bat","Tennis Racket","Yoga Mat",
    "T-Shirt","Jeans","Jacket","Shoes","Cap",
    "Rice","Milk","Coffee","Tea","Sugar"
]

categories = [
    "Electronics",
    "Furniture",
    "Stationery",
    "Sports",
    "Clothing",
    "Grocery"
]

products_data = []

for i in range(1, 51):

    products_data.append(
        (
            f"P{i:03}",
            product_names[i-1],
            random.choice(categories),
            random.randint(5, 2500),
            random.randint(10, 500)
        )
    )

product_columns = [
    "ProductID",
    "ProductName",
    "Category",
    "Price",
    "Stock"
]

products_df = spark.createDataFrame(products_data, product_columns)

products_df.coalesce(1) \
    .write \
    .mode("overwrite") \
    .option("header", True) \
    .csv(f"{BASE_PATH}/products")
```

```python
start_date = datetime(2025, 1, 1)

sales_data = []

for i in range(1, 501):

    product_id = f"P{random.randint(1,50):03}"

    quantity = random.randint(1,10)

    price = random.randint(5,2500)

    sale_amount = quantity * price

    sale_date = start_date + timedelta(days=random.randint(0,364))

    sales_data.append(
        (
            f"S{i:04}",
            random.randint(101,200),
            product_id,
            quantity,
            sale_amount,
            sale_date.strftime("%Y-%m-%d")
        )
    )

sales_columns = [
    "SaleID",
    "EmployeeID",
    "ProductID",
    "Quantity",
    "SaleAmount",
    "SaleDate"
]

sales_df = spark.createDataFrame(sales_data, sales_columns)

sales_df.coalesce(1) \
    .write \
    .mode("overwrite") \
    .option("header", True) \
    .csv(f"{BASE_PATH}/sales")
```

```python

# Create New Sample Data
# We will explicitly define the schema first so Spark knows exactly what types to use right out of the gate.

from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType

# 1. Define a clean schema
new_schema = StructType([
    StructField("EmployeeID", IntegerType(), True),
    StructField("FirstName", StringType(), True),
    StructField("LastName", StringType(), True),
    StructField("Age", IntegerType(), True),
    StructField("Salary", DoubleType(), True),
    StructField("City", StringType(), True)
])

# 2. Provide the raw data rows
new_data = [
    (101, "Alice", "Smith", 30, 85000.00, "New York"),
    (102, "Bob", "Jones", 45, 105000.50, "Chicago"),
    (103, "Charlie", "Brown", 22, 62000.00, "San Francisco"),
    (104, "Diana", "Prince", 35, 98000.25, "Seattle")
]

# 3. Create the DataFrame
df_new = spark.createDataFrame(data=new_data, schema=new_schema)
df_new.show(5, truncate=False)
```

```python
# Write to Parquet and Delta
# Now that we have df_new with a perfectly clean numeric Salary column, we can save it to both storage formats.

# Define paths
new_parquet_path = "/Volumes/workspace/default/practice-data/employees1_parquet"
new_delta_path = "/Volumes/workspace/default/practice-data/employees1_delta"

# Save as Parquet
df_new.write.mode("overwrite").parquet(new_parquet_path)
print("New data written to Parquet format!")

# Save as Delta
df_new.write.format("delta").mode("overwrite").save(new_delta_path)
print("New data written to Delta table format!")
```
