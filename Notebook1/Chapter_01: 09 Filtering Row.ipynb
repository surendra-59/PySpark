##### 1. Filtering Rows (filter() / where())

```python
# Syntax:

df.filter(condition)

# or

df.where(condition)
# Both are exactly the same.
```

```python
# Sample Data
from pyspark.sql import Row

data = [
    Row(emp_id=101, name="Alice", age=25, salary=50000, dept="HR"),
    Row(emp_id=102, name="Bob", age=30, salary=70000, dept="IT"),
    Row(emp_id=103, name="Charlie", age=35, salary=90000, dept="IT"),
    Row(emp_id=104, name="David", age=28, salary=45000, dept="Finance"),
    Row(emp_id=105, name="Eva", age=40, salary=100000, dept="HR")
]

df = spark.createDataFrame(data)

df.show()
```

```python
# Filter salary greater than 60000

from pyspark.sql.functions import col

df.filter(col("salary") > 60000).show()
```

```python
# Filter department
df.filter(col("dept") == "IT").show()
```

```python
# Multiple Conditions (AND)
df.filter(
    (col("salary") > 60000) &
    (col("age") > 30)
).show()
```

```python
# Multiple Conditions (OR)
df.filter(
    (col("dept") == "HR") |
    (col("dept") == "Finance")
).show()
```

```python
# NOT
df.filter(~(col("dept") == "HR")).show()
```

```python
# isin()
df.filter(col("dept").isin("HR","IT")).show()
```

```python
# String contains
df.filter(col("name").contains("a")).show()
```

```python
# startswith

df.filter(col("name").startswith("A")).show()
```

```python
# endswith

df.filter(col("name").endswith("e")).show()
```

##### 2. Sorting Data (orderBy() / sort())

```python
# Syntax
df.orderBy("salary")
```

```python
# Ascending
df.orderBy("salary").show()
```

```python
# Descending
from pyspark.sql.functions import col

df.orderBy(col("salary").desc()).show()
```

```python
# Multiple Columns
df.orderBy(
    col("dept"),
    col("salary").desc()
).show()


# First sort by department

# Inside each department

# Sort salary descending
```

```python
# sort()
    # Exactly same as df.orderBy()
df.sort("salary").show()
```
