```python
# Withput inference Salary is read as string
df2 = spark.read.csv(
    "/Volumes/workspace/default/practice-data/employees/",
    header=True,
    inferSchema=True
)
```

```python
df2.printSchema()
# When Spark scans a CSV file to infer types, it evaluates the data from top to bottom. It follows a strict hierarchy: it will try to match Integer, Long, or Double first, but if even one single row contains a non-numeric character, Spark forces the entire column to fall back to StringType to avoid data loss.
```

```python
# Read Parquet
df3 = spark.read.parquet(
    "/Volumes/workspace/default/practice-data/employees1_parquet/"
)
```

```python
df3.printSchema()

# You don't need inferSchema=True for Parquet because Parquet files inherently store their own schema inside the file itself.

# When you read a CSV, it's just raw, unformatted text. Spark has to scan the characters and guess what they are.

# But a Parquet file is a smart binary format. It contains a section at the very end of the file called the footer. This footer holds metadata, which explicitly lists:

# The exact column names

# The exact data types (Integer, Double, String, etc.)

# Even basic statistics (like minimum and maximum values for each column)

# So, when you run spark.read.parquet(), Spark doesn't have to guess or scan the whole dataset. It simply jumps to the footer, reads the embedded schema definition, and instantly knows that Salary is a double and Age is an integer.

# Delta tables work the exact same way—they keep a strict transaction log (_delta_log) that spells out the exact datatypes, so you never have to guess or use inferSchema again!
```
