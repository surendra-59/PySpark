#### Chapter 1 – Topic 5: Understanding Schema (PySpark)

What is a Schema?

A schema defines the structure of a DataFrame.

It tells Spark:

Column names
Data types
Nullable (Yes/No)

##### Why Schema Matters
######Without a schema Spark doesn't know:
Which columns exist, 
Which datatype each column has, 
How to optimize queries

######Correct schema means:
Faster execution,
Less memory usage,
Better query optimization,
Fewer errors

```python
df = spark.read.csv(
    "/Volumes/workspace/default/practice-data/sales/",
    header=True,inferSchema=True
)
```

```python
# View Schema
df.printSchema()
```

```python
# Common Data Types       
# Type	         Meaning     
# StringType	     Text        
# IntegerType	     Whole numbers       
# LongType	         Large integers      
# DoubleType	     Decimal numbers     
# FloatType	         Decimal numbers     
# BooleanType	     True/False      
# DateType	         Date  
# TimestampType	     Date & Time  
# ArrayType	         List  
# StructType	     Nested object  
# MapType	         Key-Value  
```

```python
# Infer Schema Automatically
# Spark tries to detect datatypes automatically.
df1 = (
    spark.read
    .option("header",True)
    .option("inferSchema",True)
    .csv("/Volumes/workspace/default/practice-data/employees/")
)
df1.printSchema()
```

```python
df1.display()
```

```python
# Without inferSchema
# Everything becomes String
df2 = (
    spark.read
    .option("header",True)
    .csv("/Volumes/workspace/default/practice-data/employees/")
)
```

```python
# Manually Defining Schema (Recommended)
from pyspark.sql.types import *

schema = StructType([
    StructField("SalesID", StringType(), True),
    StructField("EmployeeID", IntegerType(), True),
    StructField("ProductID", StringType(), True),
    StructField("Quantity", IntegerType(), True),
    StructField("SaleAmount", IntegerType(), True),
    StructField("SaleDate", DateType(), True),
])

df3 = spark.read.schema(schema).csv("/Volumes/workspace/default/practice-data/sales/", header=True) 
```

```python
display(df3.limit(5))
```

```python
# Advantages of Manually Defining Schema
    # Faster than inferSchema
    # More reliable
    # Prevents incorrect type inference
    # Preferred in production
```
